Matplotlib commands are nearly identical to Matlab commands. So, in case you know Matlab, you won't have to learn much.

```python
# %pylab inline

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (5,3.5)
```

```python
# Generate some test data
x = np.arange(0, 6.28, 0.01)
y = np.sin(x)
```

Now we plot y versus x

```python
plt.plot(x, y)
```

## Adding Labels

That was simple: we have plotted y as a function of x. Now let us add axis labels and a title.

```python
plt.plot(x, y)
plt.xlabel("theta (angle in radians)")
plt.ylabel('sin(theta)')
plt.title("This is my first plot")
```

## Fancier labels:

```python
plt.plot(x, y)
xlab = plt.xlabel(r"$\theta$ (angle in radians)") # note the r before quotes
ylab = plt.ylabel(r'sin($\theta$)')
thetitle= plt.title("This is is a nicer plot");thetitle.set_fontweight('bold')
ylab.set_color('blue');xlab.set_color('red'); xlab.set_style('italic')
```

More examples at https://matplotlib.org/stable/gallery/text_labels_and_annotations/fonts_demo.html

## Axis Range and scale

```python
# Let's make it look a bit better
plt.plot(another_x,another_x**4)
plt.title(r'X$^4$ versus X') # The $ signs are for latex inputs
plt.xlim(min(another_x), max(another_x))
plt.yscale('log')
plt.show()
```

## Adding a Legend

```python
plt.plot(x, y, label='Sine(x)')
plt.plot(x, np.cos(x))
y_scatter = y+np.random.normal(0.0, 0.1, len(x))
plt.plot(x, y_scatter, ls='None', marker='.', label='Fake data with scatter')
```

*np.random.normal()* generated len(x) data points with mean 0.0 and sigma 0.1.  

```python
plt.plot(x, y, label='Sine(x)'); plt.plot(x, np.cos(x))
y_scatter = y+np.random.normal(0.0, 0.1, len(x))
plt.plot(x, y_scatter, ls='None', marker=',', label='Fake data with scatter')
plt.legend()
```

Note that we did not add a "label" keyword in the cos(x) plot, so that is not listed in the legend

## Error bars

```python
y_error = np.random.normal(0.0, 0.2, len(x_spaced))
x_error = np.repeat(0.1, len(x_spaced))
plt.errorbar(x_spaced, np.sin(x_spaced), yerr=y_error,xerr=x_error,marker='s',ecolor='r')
```

## Autoscaling of axes

```python
mjd = [56705.7130824, 56705.7245194, 56705.7558403, 56705.7591719, 56705.7784380,56705.8334423, 
       56705.9293922, 56705.9611691, 56705.9814212,  56706.0066355]
# continuing on next line is okay
magnitude = [18.03, 18.32, 17.89, 18.12, 18.12, 17.97, 18.44, 18.28, 18.25, 18.61]
magerr = [0.13, 0.10, 0.08, 0.13, 0.13, 0.14, 0.05, 0.10, 0.09, 0.08]
plt.errorbar(mjd, magnitude, yerr=magerr,color='#ff8800',lw=3)
```

Note the X axis: Matplotlib has automatically subtracted a constant and made the plot more readable.

```python
plt.errorbar(mjd, magnitude, yerr=magerr, 'ro') # does not work
```

```python
# instead of shorthand, give the full command
plt.errorbar(mjd, magnitude, yerr=magerr, marker='o', color='r', ecolor='g', ls='None')
```

Note: Y axis is still flipped. Higher magnitude is fainter, so it should be plotted below.

```python
plt.errorbar(mjd, magnitude, yerr=magerr, marker='o', color='#ff9900', ecolor='g', ls='None')
ymin, ymax = plt.ylim()
plt.ylim(ymax, ymin)
plt.ticklabel_format(useOffset=False)
plt.title("Y axis flipped, X label offset removed")
```

X axis labels: the offset is not subtracted anymore, but now labels are hard to read.

```python
plt.errorbar(mjd, magnitude, yerr=magerr, marker='o', color='r', ecolor='g', ls='None')
ymin, ymax = plt.ylim()
plt.ylim(ymax, ymin)
plt.ticklabel_format(useOffset=False)
plt.xticks(rotation=45)
plt.title("X labels rotated to become more readable")
```

Subplots
===

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.rcParams['figure.figsize'] = (3,3)
plt.subplot(221); plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(222); plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.subplot(223); plt.plot(t2,np.sin(2.*np.pi*t2),'g-.')
plt.subplot(224); plt.plot(t2,t2**2,'r-')

plt.savefig('subplots.png'); plt.savefig('subplots.eps'); plt.savefig('subplots.svg')
```

Contour plots
===

```python
def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)
n = 256; x = np.linspace(-3, 3, n); y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
plt.axes([0.025, 0.025, 0.95, 0.95])
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidths=.5)
plt.clabel(C, inline=1, fontsize=20)
```

2D Histogram
===

```python
# Generate sample data
n_stars = 10000
color = np.random.normal(0.6, 0.2, n_stars)
magnitude = 3*color + np.random.normal(15, 0.5, n_stars)
rgb_stars = int(0.2 * n_stars)
rgb_color = np.random.normal(1.2, 0.1, rgb_stars)
rgb_magnitude = np.random.normal(13, 0.8, rgb_stars)
color = np.concatenate([color, rgb_color])
magnitude = np.concatenate([magnitude, rgb_magnitude])

# Create 2D histogram
plt.figure(figsize=(6, 4.5))
plt.hist2d(color, magnitude, bins=50, cmap='viridis')
plt.colorbar(label='Number of Stars')

plt.gca().invert_yaxis()
plt.xlabel('B-V Color')
plt.ylabel('Absolute Magnitude')
plt.title('Dummy Stellar Color-Magnitude Diagram')
```

3D plotting with mplot3d
===

```python
#See: https://matplotlib.org/stable/gallery/mplot3d/surface3d_radial.html
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5, 3.5)) # OOP syntax
ax = fig.add_subplot(111, projection='3d')
r = np.linspace(0,1.25,50) # create supporting points in polar coordinates
p = np.linspace(0,2*np.pi,50)
R,P = np.meshgrid(r,p)
X,Y = R*np.cos(P),R*np.sin(P) # transform them to cartesian system
Z = ((R**2 - 1)**2)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
```

# Statistical Visualization with Seaborn
## Why Seaborn?
- Built on matplotlib with statistical focus
- Excellent for distribution visualization
- Beautiful default styles
- Great for large astronomical datasets
- Simple interface for complex plots

```python
# Basic Seaborn Setup
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better astronomy plots
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 5))

# Generate sample star cluster data
n_stars = 200
magnitude = np.random.normal(15, 2, n_stars)
color = np.random.normal(0.6, 0.3, n_stars)

```

```python
# Create scatter plot with density estimation
sns.scatterplot(x=color, y=magnitude)
plt.gca().invert_yaxis()  # Astronomical convention
plt.xlabel('Color Index (B-V)')
plt.ylabel('Apparent Magnitude')
plt.title('Dummy Star Cluster CMD with Seaborn')
```

## Key Seaborn Plot Types for Astronomy
- scatterplot: CMD diagrams, correlation plots
- histplot: magnitude distributions
- kdeplot: density estimation
- jointplot: correlations with distributions
- heatmap: correlation matrices, 2D histograms

```python
# Distribution Visualization
plt.figure(figsize=(8, 5))

# Simulate galaxy redshift data
redshifts = np.concatenate([
    np.random.normal(0.5, 0.1, 300),
    np.random.normal(1.2, 0.2, 200)
])

# Plot distribution with KDE
sns.histplot(data=redshifts, kde=True)
plt.xlabel('Redshift (z)')
plt.ylabel('Count')
plt.title('Galaxy Redshift Distribution')
```

```python
# Joint Distribution Example
# Good for showing correlations and distributions together
stellar_data = {
    'temp': 5000 + np.random.normal(0, 500, 100),
    'luminosity': np.random.lognormal(0, 0.5, 100)
}

sns.jointplot(
    data=stellar_data,
    x='temp', y='luminosity',
    kind='scatter',
    marginal_kws=dict(bins=20)
)
plt.subplots_adjust(top=0.95)
plt.suptitle('Dummy Stellar Temperature-Luminosity Relation')
```

```python
# Correlation Heatmap Example
# Useful for multi-wavelength studies
obs_data = {
    'UV': np.random.normal(0, 1, 50),
    'optical': np.random.normal(0, 1, 50),
    'IR': np.random.normal(0, 1, 50),
    'radio': np.random.normal(0, 1, 50)
}
import pandas as pd
df = pd.DataFrame(obs_data)
correlation = df.corr()

plt.figure(figsize=(6, 4.5))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Dummy Multi-wavelength Correlation Matrix')
```

```python
# Density Plot Example with Seaborn
import seaborn as sns

# Set style and figure size
sns.set_theme(style="darkgrid")
plt.figure(figsize=(5, 3))

# Generate sample data: simulating star cluster metallicities
# Two populations: disk and halo stars
n_stars = 1000
disk_stars = np.random.normal(-0.2, 0.2, int(0.7 * n_stars))  # [Fe/H] for disk
halo_stars = np.random.normal(-1.5, 0.4, int(0.3 * n_stars))  # [Fe/H] for halo
metallicities = np.concatenate([disk_stars, halo_stars])

# Create density plot
sns.kdeplot(data=metallicities, fill=True)
plt.xlabel('[Fe/H]')
plt.ylabel('Density')
plt.title('Dummy Metallicity Distribution of Stars')

# Add vertical lines for population means
plt.axvline(x=-0.2, color='r', linestyle='--', alpha=0.5, label='Disk Mean')
plt.axvline(x=-1.5, color='b', linestyle='--', alpha=0.5, label='Halo Mean')
plt.legend()
```
